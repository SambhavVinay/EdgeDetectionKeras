```python
# Open in Colab
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras import layers, Sequential
from google.colab import files

# Upload an image
uploaded = files.upload()

for fn in uploaded.keys():
    # Load and preprocess image
    img = image.load_img(fn, target_size=(150, 150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    # Define vertical and horizontal edge detection filters (Sobel-like)
    vertical_edge = np.array([[-1., 0., 1.],
                              [-2., 0., 2.],
                              [-1., 0., 1.]], dtype=np.float32)

    horizontal_edge = np.array([[-1., -2., -1.],
                                [ 0.,  0.,  0.],
                                [ 1.,  2.,  1.]], dtype=np.float32)

    # Reshape to Conv2D kernel format (3,3,in_channels,out_channels)
    vertical_edge = vertical_edge.reshape((3,3,1,1))
    horizontal_edge = horizontal_edge.reshape((3,3,1,1))

    # Repeat for 3 input channels (RGB)
    vertical_edge = np.repeat(vertical_edge, 3, axis=2)
    horizontal_edge = np.repeat(horizontal_edge, 3, axis=2)

    # Build models
    vertical_model = Sequential([
        layers.Conv2D(1, (3,3),
                      kernel_initializer=tf.constant_initializer(vertical_edge),
                      bias_initializer='zeros',
                      padding='same',
                      input_shape=(150,150,3),
                      trainable=False)
    ])

    horizontal_model = Sequential([
        layers.Conv2D(1, (3,3),
                      kernel_initializer=tf.constant_initializer(horizontal_edge),
                      bias_initializer='zeros',
                      padding='same',
                      input_shape=(150,150,3),
                      trainable=False)
    ])

    # Apply filters
    vert_edges = vertical_model.predict(img_array)
    horiz_edges = horizontal_model.predict(img_array)

    # Combine edges (magnitude)
    combined_edges = np.sqrt(np.square(vert_edges) + np.square(horiz_edges))

    # Plot results
    plt.figure(figsize=(14,4))

    plt.subplot(1,4,1)
    plt.imshow(img_array[0])
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1,4,2)
    plt.imshow(vert_edges[0].squeeze(), cmap="gray")
    plt.title("Vertical Edges")
    plt.axis("off")

    plt.subplot(1,4,3)
    plt.imshow(horiz_edges[0].squeeze(), cmap="gray")
    plt.title("Horizontal Edges")
    plt.axis("off")

    plt.subplot(1,4,4)
    plt.imshow(combined_edges[0].squeeze(), cmap="gray")
    plt.title("Combined Edges")
    plt.axis("off")

    plt.show()
```
